```python
# AlgoSage Experimental Metrics Analysis
# Author: [Kavyasruthi V S]
# Title: AlgoSage: An AI Mentor for Real-Time Algorithm Optimization in Software Development

import json
import pandas as pd
import matplotlib.pyplot as plt

# Load metrics from JSON
with open('metrics.json', 'r') as file:
    data = json.load(file)

# Convert JSON data to DataFrame for better handling
df = pd.DataFrame([data])

# Display the metrics summary
print("=== AlgoSage Experimental Metrics Summary ===")
print(df.T)

# Plot selected metrics for visual presentation
metrics_to_plot = ['recommendation_accuracy', 'precision', 'recall', 'optimization_efficiency']
plt.figure(figsize=(6, 6))
plt.bar(metrics_to_plot, df[metrics_to_plot].iloc[0], width=0.5)
plt.title('AlgoSage Performance Evaluation Metrics', fontsize=13)
plt.ylabel('Score / Percentage')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Derived interpretation metrics
efficiency_gain = df['optimization_efficiency'].iloc[0] * 100
print(f"\nAverage Optimization Efficiency: {efficiency_gain:.2f}%")
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# Load metrics from your file
with open("metrics.json", "r") as file:
    data = json.load(file)

# Convert to DataFrame
df = pd.json_normalize(data)

# Print column names (to verify structure)
print("Columns available in JSON:\n", df.columns.tolist(), "\n")

# Extract the correct metric columns (matching your file)
metrics = {
    "Recommendation Accuracy": float(df["Recommendation Accuracy"].iloc[0]),
    "Precision": float(df["Precision"].iloc[0]),
    "Recall": float(df["Recall"].iloc[0]),
    "Optimization Efficiency (%)": float(df["Optimization Efficiency (simulated avg %)"].iloc[0])
}

# Create DataFrame for plotting
plot_df = pd.DataFrame(list(metrics.items()), columns=["Metric", "Value"])

# Create IEEE-style bar chart
plt.figure(figsize=(6,6))
plt.bar(plot_df["Metric"], plot_df["Value"],
        color=['steelblue','darkorange','seagreen','crimson'], width=0.5)
plt.title("AlgoSage Performance Evaluation Metrics", fontsize=14, fontweight="bold")
plt.ylabel("Score / Percentage", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()

# Save graph for your paper
plt.savefig("AlgoSage_Performance_Graph.png", dpi=300)
plt.show()
```
